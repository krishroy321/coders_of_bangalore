*Data Collection*

```python
with open ("data.txt",'r',encoding = 'utf-8') as f:
    data = f.read()
```

```python
chunks = data.split("\n\n")
chunks = [c for c in chunks if len(c)>3]
```

```python
def parse_chunk(chunk):
    try:
        chunk = chunk.strip()
        sep_chunk = chunk.split("\n")
        username = sep_chunk[0] if len(sep_chunk) > 0 else "" 
        name = sep_chunk[1] if len(sep_chunk) > 1 else ""
        no_of_posts = int(sep_chunk[2].split(" post")[0])if len(sep_chunk) > 2 else ""
        no_of_followers = float(sep_chunk[3].split(" follower")[0].replace(",","").replace(".","").replace("K","").replace("M","")if len(sep_chunk) > 3 else "")
        if ("K" in sep_chunk[3]):
            no_of_followers = int(no_of_followers * 1000)
        elif ("M" in sep_chunk[3]):
            no_of_followers = int(no_of_followers * 1000000)
        else:
            no_of_followers = int(no_of_followers)

        no_of_following = float(sep_chunk[4].split(" following")[0].replace(",","").replace(".","").replace("K","000").replace("M","000000")if len(sep_chunk) > 4 else "")
        if ("K" in sep_chunk[4]):
            no_of_following = int(no_of_following * 1000)
        elif ("M" in sep_chunk[4]):
            no_of_following = int(no_of_following * 1000000)
        else:
            no_of_following = int(no_of_following)
        
        if len(sep_chunk)>5:
            type_of_page = sep_chunk[5]
            bio = "\n".join(sep_chunk[6:]).replace("\n"," ")
        else:
            type_of_page = "Unknown"
            bio = ""
        return {"Username":username, 
                "Name":name,
                "No of Post":no_of_posts,
                "No of Followers":no_of_followers,
                "No of Following":no_of_following,
                "Type Of Page ":type_of_page,
                "Bio ":bio}
    except Exception as e:
        print("Error:",e)
        print(chunk)
```

```python
parse_chunk(chunks[49])
```

```python
all_chunks = []
for chunk in chunks:
    all_chunks.append(parse_chunk(chunk))
print(all_chunks[46])
```

```python
import json
s = json.dumps(all_chunks,indent=4,ensure_ascii=False)
print(s)
```

Who Has The Maximum Posts?

```python
max_post = 0
for chunk in all_chunks:
    if (max_post<chunk['No of Post']):
        max_post = chunk['No of Post']
        chunk_with_max_post = chunk
print(chunk_with_max_post)
```
